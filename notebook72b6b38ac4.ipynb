## 1. Data Cleaning & Auditing (SQL)

-- Checking for duplicate orders to ensure data accuracy
SELECT `Order ID`, `Product ID`, COUNT(*)
FROM `hardy-messenger-472122-c1.my_project.superstore_sales`
GROUP BY `Order ID`, `Product ID`
HAVING COUNT(*) > 1;

-- Verifying annual sales performance to align with the dashboard
SELECT 
    EXTRACT(YEAR FROM Order_Date) AS Year, 
    SUM(Sales) AS Annual_Sales
FROM `hardy-messenger-472122-c1.my_project.superstore_sales`
GROUP BY Year
ORDER BY Year;
